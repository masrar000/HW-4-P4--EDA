```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = "/workspaces/codespaces-jupyter/titanic.csv"
data = pd.read_csv(url)
```

```python
print(data.head())
```

```python
for column in data.columns:
    plt.figure(figsize=(10, 5))
    
    # If data type is object, consider it categorical
    if data[column].dtype == 'object':
        plt.xticks(rotation=45)  # Rotate x-axis labels for readability
        
        if column == 'Name' or column == 'Ticket' or column == 'Cabin':
            sns.countplot(data=data, y=column, order=data[column].value_counts().index[:10])
        else:
            sns.countplot(data=data, y=column)
            
        plt.title(f'Bar chart of {column}')
    else:
        sns.histplot(data[column], kde=True)
        plt.title(f'Histogram of {column}')
        # Reporting
        print(f"\nColumn: {column}")
        print("Max:", data[column].max())
        print("Min:", data[column].min())
        print("Mean:", data[column].mean())
        print("Median:", data[column].median())
        print("Outliers:", [outlier for outlier in data[column] if outlier > data[column].mean() + 3 * data[column].std() or outlier < data[column].mean() - 3 * data[column].std()])
    
    plt.show()


```

```python
plt.figure(figsize=(20,10))
sns.boxplot(data=data.select_dtypes(include=['float64', 'int64']))
plt.xticks(rotation=45)
plt.title('Boxplot of All Numeric Variables')
plt.show()
```

```python
sns.pairplot(data.select_dtypes(include=['float64', 'int64']))
plt.show()
```

```python
cross = pd.crosstab(data['Survived'], data['Embarked'])
sns.heatmap(cross, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Using boxplot
sns.boxplot(x=data['Pclass'], y=data['SibSp'])
plt.show()

# Using scatter plot with color
sns.scatterplot(x=data['Parch'], y=data['Fare'], hue=data['Sex'])
plt.show()
```

```python
#Categorical Columns:

'''Survived: Indicates whether the passenger survived (1) or did not survive (0).
Pclass: The class of ticket the passenger purchased (1st, 2nd, or 3rd).
Name: The name of the passenger.
Sex: The gender of the passenger (male or female).
Ticket: The ticket number.
Cabin: The cabin number where the passenger stayed.
Embarked: The port where the passenger boarded the Titanic. C represents Cherbourg, Q represents Queenstown, and S represents Southampton.'''
```

```python
#Numerical Columns:

'''Age: The age of the passenger.
SibSp: The number of siblings or spouses the passenger had aboard the Titanic.
Parch: The number of parents or children the passenger had aboard the Titanic.
Fare: The amount of money the passenger paid for the ticket.'''
```
